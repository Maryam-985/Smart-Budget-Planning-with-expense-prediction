```python
# 2. FIX ZERO AMOUNT ISSUE
# Replace amount=0 with transfer-amount
df['amount'] = df.apply(
    lambda row: row['transfer-amount'] if row['amount'] == 0 else row['amount'],
    axis=1
)
```

```python
df.head(64)
```

```python
# 3. SELECT REQUIRED COLUMNS
df.columns = df.columns.str.strip().str.lower()
df = df[['date', 'amount', 'category','account', 'title', 'type']]
```

```python
print(df['account'].unique())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# fill amount nulls with 0
df['amount'] = df['amount'].fillna(0)
```

```python
# fill category nulls
df['category'] = df['category'].fillna("not defined")
```

```python
# fill title nulls using account
df['title'] = df['title'].fillna(df['account'])
```

```python
# 5. DATE CONVERSION
df['date'] = pd.to_datetime(df['date'])
```

```python
# 6. CLEAN TEXT DATA

df['title'] = df['title'].str.lower()
df['category'] = df['category'].str.lower()

df['title'] = df['title'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)

df['title'] = df['title'].str.strip()
df['category'] = df['category'].str.strip()
```

```python
df.head(5)
```

```python
# 7. FILTER ONLY EXPENSES
df = df[df['type'].str.lower() == 'expense']
# 8. REMOVE INVALID AMOUNTS
df = df[df['amount'] > 0]
```

```python
# =========================
# 9. FEATURE ENGINEERING
# =========================
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.weekday
```

```python
df.head(5)
```

```python
# =========================
# 10. RESET INDEX
# =========================
df = df.reset_index(drop=True)
```

```python
# =========================
# 11. SAVE CLEAN DATA
# =========================
df.to_csv("cleaned_expenses.csv", index=False)
```
